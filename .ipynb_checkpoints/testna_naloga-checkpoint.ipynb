```python
def standardize(x):
    return (x - x.mean()) / x.std()

# get X with Y, where X is not NaN
orig_sig = df['X'][df['X'] == df['X']]
#removing one sample from X to make the orig_sig odd length, makes sliding window operations easier
orig_sig = orig_sig[1:]
orig_sig = orig_sig.reset_index(drop=True)
# standardizing it
orig_sig = standardize(orig_sig)

len(orig_sig)
print(orig_sig)
```

```python
import scipy.optimize
import numpy as np

epsilon = 0.00001
orig_sig_len = len(orig_sig)
mean_factor = 1.0 / orig_sig_len

# returns as minimal as possible L2 norm
def min_errorL2(x, y):
    
    # kill the almost zero elements
    #y[np.abs(y) < 0.000001] = 0
    
    # x and y are vectors of same length
    # alpha is projection length, solution to min(theta) ||x-theta*y||
    # alpha minimizes norm
    alpha = (x*y).sum() / ( (y*y).sum() + epsilon)
    
    # difference
    r = (x - (alpha * y))
    
    # return the calculated L2 norm and alpha, also calculates L1 norm
    return (np.sqrt(mean_factor * (r*r).sum()), alpha, mean_factor * (np.abs(r)).sum())

print(min_errorL2(np.array([1,0,1]), np.array([2,0,-2])))

def rbf(r):
    s = np.sign(r)
    r = np.abs(r)
    return s / (1.0 + r*r)

# lets try custom windowing
def signal_compare_L2(window):    
    
    result = min_errorL2(orig_sig, window)
    return rbf(result[0])# * np.log(np.abs(result[1]) + epsilon)

def signal_compare_L1(window):    
    
    result = min_errorL2(orig_sig, window)
    return rbf(result[2])# * np.log(np.abs(result[1]) + epsilon)
    
# test derivative numerically
#scipy.optimize.check_grad(errorL1, grad_errorL1, np.array([0.0, 1.0, 1.0]), *(orig_sig, orig_sig))
```

```python
def rms(window):
    
    return np.sqrt((window*window).mean())

# cross correlation
def corr(window):
    
    return (orig_sig*window).mean()
```

```python
from scipy.stats.stats import pearsonr

# pearson
def pearson(window):
    
    result = pearsonr(orig_sig, window)
    #print(result)
    return result[0]
```

```python
diff_rolling = pd.DataFrame()

diff_rolling['ORIG'] = df['X'] 
#diff_rolling['ORIG_DB'] = 20 * np.log(df['X'])
diff_rolling['SIG'] = df['Y']
diff_rolling['SIG_RMS'] = df['Y'].rolling(orig_sig_len, center=True).apply(rms, raw=True)
diff_rolling['MIN_L2'] = df['Y'].rolling(orig_sig_len, center=True).apply(signal_compare_L2, raw=True)
diff_rolling['MIN_L1'] = df['Y'].rolling(orig_sig_len, center=True).apply(signal_compare_L1, raw=True)
diff_rolling['MIN_L2_RMS'] = diff_rolling['MIN_L2'].rolling(orig_sig_len, center=True).apply(rms, raw=True)

diff_rolling['CORR'] =  df['Y'].rolling(orig_sig_len, center=True).apply(corr, raw=True)
diff_rolling['CORR_RMS'] = diff_rolling['CORR'].rolling(orig_sig_len, center=True).apply(rms, raw=True)

diff_rolling['MIN_L2*CORR'] = diff_rolling['MIN_L2'] * diff_rolling['CORR'].abs()

diff_rolling['MIN_L2_RMS*CORR_RMS'] = diff_rolling['MIN_L2_RMS'] * diff_rolling['CORR_RMS'].abs()

diff_rolling['PEARSON'] = df['Y'].rolling(orig_sig_len, center=True).apply(pearson, raw=True).abs()
diff_rolling['PEARSON_RMS'] = diff_rolling['PEARSON'].rolling(orig_sig_len, center=True).apply(rms, raw=True)
```

```python
import scipy.fftpack
```

```python
orig_df = pd.DataFrame()

orig_df['ORIG']      = orig_sig
orig_df['ORIG_FFT']  = scipy.fftpack.fft(orig_sig) * (1/orig_sig_len)
orig_df['ORIG_FFT_REAL'] = orig_df['ORIG_FFT'].real
orig_df['ORIG_FFT_IMAG'] = orig_df['ORIG_FFT'].imag
orig_df[['ORIG_FFT_REAL', 'ORIG_FFT_IMAG']].plot()
```

```python
half_win = orig_sig_len//2

fft_window = pd.DataFrame()

for window_center in range(half_win, len(df['Y']) - half_win):
    win_start = window_center - half_win
    win_end = window_center + half_win
    
    window = df['Y'][win_start:win_end]
    # there are complex numbers
    result = scipy.fftpack.fft(window) * (1/orig_sig_len)
    fft_window[window_center] = np.abs(result)
```

```python

def fft_dot_product(window):
    fft_window = scipy.fftpack.fft(window) * (1.0/orig_sig_len)
    product = np.dot(fft_window, orig_df['ORIG_FFT'])
    return np.abs(product)

# returns as minimal as possible L2 norm
def min_errorL2_complex(x, y):
    
    # x and y are vectors of same length
    # alpha is projection length, solution to min(theta) ||x-theta*y||
    # alpha minimizes norm
    alpha = (x*np.conjugate(y)).sum() / ( (y*np.conjugate(y)).sum() + epsilon)
    
    
    #print(alpha)
    
    # difference
    r = (x - (alpha * y))
    r = np.abs(r)
    # 
    return (np.sqrt(mean_factor * (r*r).sum()), alpha)

def fft_min_L2(window):
    
    fft_window = scipy.fftpack.fft(window) * (1.0/orig_sig_len)
    
    result = min_errorL2_complex(orig_df['ORIG_FFT'], fft_window)
    
    return rbf(result[0])# * np.log(np.abs(result[1]) + epsilon)
```

```python
diff_rolling['FFT'] = df['Y'].rolling(orig_sig_len, center=True).apply(fft_dot_product, raw=True)
diff_rolling['FFT_MIN_L2'] = df['Y'].rolling(orig_sig_len, center=True).apply(fft_min_L2, raw=True)
```

```python


def find_max_signal(signal, sample, method = 'MIN_L2'):

    half_len = len(sample) // 2    
    
    if method == 'MIN_L2':        
        # lets try custom windowing
        def min_L2_func(window):
            
            result = min_errorL2(sample, window)
            return result[0]
        
        result = signal.rolling(len(sample), center = True).apply(min_L2_func, raw = True)
        
        # find the smallest L2 value
        index = result.idxmin()
        
        subset = signal[index-half_len:index+half_len+1].copy().reset_index(drop=True)
       
        error = min_errorL2(subset, sample)
        
        return index, error
    
    raise Exception("unknown method")
```

```python
find_max_signal(df['Y'], orig_sig)
```

```python
iterations = pd.DataFrame()

temp = df['Y'].copy()

iterations[0] = temp

for i in range(1,6):

    (index, max_sig) = find_max_signal(temp, orig_sig)

    for k in range(0, orig_sig_len):
        temp[index - half_len + k] = temp[index - half_len + k] - max_sig[1] * orig_sig[k]

    iterations[i] = temp.copy()
    
```
